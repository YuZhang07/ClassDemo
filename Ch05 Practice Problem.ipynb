```python
# List all Female Senior Game Designer's ID, first name, last name
# And: use &

# Understand steps
# 1) identify the rows
res1 = (data['Title'] =='Senior Programmer') & (data['Gender'] =='F')
#print(res1)

# 2) filter the rows
res2 = data[res1]

# 3) only display ID, first name, last name
res3 = res2[['ID','Last Name','First Name','Title']]
res3


# one statement
data[(data['Title'] =='Senior Programmer') & (data['Gender'] =='F')][['ID','Last Name','First Name','Title']]
```

```python
# List all Senior Sound Designer or Senior Game Designer
# OR : use |
res = data[(data['Title'] =='Senior Programmer') | (data['Title'] =='Senior Sound Designer') ]


# List all salary of Senior Sound Designer or Senior Game Designer
res.Salary

# List average salary of Senior Sound Designer or Senior Game Designer
print(res.Salary.mean())

# One statement
data[(data['Title'] =='Senior Programmer') | (data['Title'] =='Senior Sound Designer') ].Salary.mean()
```
