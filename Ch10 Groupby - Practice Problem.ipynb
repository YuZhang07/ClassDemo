```python
# Compare the average salary of female employees in each title
df = data

# 1) Filter out female
res1 = df[ df.Gender == 'F']
res1

# 2) Group by department
res2 = res1.groupby(df.Department)
res2

# 3) Get average salary 
res3 = res2.Salary.mean()
res3

# 4 ) Sorting by
res4 = res3.sort_values(ascending= False)
res4.plot(kind ='bar', rot=60, title ='Female Salary By Department')
```

```python
# Group by Department, default is sort by index

res5 = data['ID'].groupby(data.Department).count()
print(res5)
res5.plot(kind='bar', title ='Employee Numbers By Department')
```

```python
# Group by Department, default is sort by index, can chance sorting by value

res5 = data['ID'].groupby(data.Department).count().sort_values()
res5 = data['ID'].groupby(data.Department).count().sort_values(ascending = False)
print(res5)
res5.plot(kind='bar', title ='Employee Numbers By Department')
```

```python
# Group by Performance, show number of person

# 1ï¼‰ Groupby
res1 = data.groupby(data.Performance)

# 2) Select ID
res2 = data.groupby(data.Performance).ID

# 3) Get ID.count
res3 = data.groupby(data.Performance).ID.count()
res3

# 4) Sorting, sort_index  vs sort_values   / ascending = True vs ascending = False
res4 = res3.sort_values(ascending = False)
print(res4)
res4.plot(kind='bar', title = 'Employee Performance')
```

```python
res4.plot(kind='pie', title='Employee Performance',autopct='%1.1f%%')
```

# Groupby two columns

```python
# Groupby two columns: data.Department, data.Gender
# Count number

# Step 1: groupby
res1 = data.groupby([data.Department, data.Gender])
res1

# Step 2: after groupby, select ID
res2 = res1.ID

# Step 3: apply a fuction
res3 = res2.count()
print(res3)

# Step 4 Sort: sort_index vs sort_values()
res4 = res3.sort_values(ascending=False)

# Step 5: plot
res4.plot(kind='bar', title= 'Employee number by Department and Gender', rot=45, color='g')
```

```python
res4.plot(kind='barh')
```

```python
print(res4)
print('-----------')
print(res4.index)
print(res4.values)
print('-----------')
print(res4.index[0])
print(res4.values[0])
print('-----------')
print(res4.index[0][0])
print(res4.index[0][1])
print(res4.index[1][0])
print(res4.index[1][1])
```

```python
# Use unstack for department and gender
res5 = res4.unstack()
res5
```

```python
# Plot number of employees by department and gender
res5.plot(kind='barh', title='employees by department and gender')
```

```python
# Unstack vs. stack
res = data.groupby([data.Title, data.Gender]).size()
print(res)
print('\n---')
res = res.unstack()
print(res)

res = res.stack()
print('\n---')
print(res)
```
